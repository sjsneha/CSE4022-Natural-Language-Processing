```python
from urllib import request
```

```python
# Task: Scrape Data from Web
url="https://www.gutenberg.org/files/98/98-0.txt"
```

```python
response=request.urlopen(url)
```

```python
raw=response.read().decode('utf8')  //reading after decoding
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens=word_tokenize(raw)
```

```python
print(tokens[:200])
```

```python
#PorterStemmer
import nltk
from nltk.stem import PorterStemmer
porter=PorterStemmer()
porter.stem('happiness')
```

```python
#RegexpStemmer
#RE or Lemmatizers might work with words like 'cacti'
import nltk
from nltk.stem import RegexpStemmer
regex=RegexpStemmer('ing')
regex.stem('walking'))
```

```python
#SnowballStemmer
import nltk
from nltk.stem import SnowballStemmer
snow=SnowballStemmer('french')
snow.stem('manges') #Wrong (Should be mange)
```

```python
snow.stem('neuf') #Correct
```

```python
snow.stem('robuste') #Correct
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
print(lemma.lemmatize("cacti"))
```

```python
print(lemma.lemmatize("mice"))
```
